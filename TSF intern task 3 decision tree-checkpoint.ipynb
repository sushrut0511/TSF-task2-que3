```python
# import relevant lib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn import tree
```

```python
#import data set
raw_data=pd.read_csv(r'C:\Users\LENOVO\Downloads\Iris.csv')
data=raw_data.copy()
data
```

```python
data=data.drop(columns=['Id'])
data
```

```python
data['Species'].unique()
```

```python
data['Species']=data['Species'].map({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})
data
```

```python
df_features=data.iloc[ : ,0:4]
df_features
```

```python
#split the data
x_train, x_test, y_train, y_test = train_test_split(df_features, data['Species'], random_state=0)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(max_depth = 3,random_state = 0)
dtc.fit(x_train, y_train)
```

```python
dtc.predict(x_test)
```

```python
tree.plot_tree(dtc)
```

```python
feature=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']
classes=['setosa', 'versicolor', 'virginica']
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (2,2), dpi=300)
tree.plot_tree(dtc,
               feature_names = feature, 
               class_names=classes,rotate=True,
               filled=True);
```
